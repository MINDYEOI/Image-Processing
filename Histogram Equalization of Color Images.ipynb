## Description

1. openCV의 `cvtColor`를 이용해 HSV 형식으로 변환하고, split을 사용해 색상, 채도, 명도값을 따로 저장
2. 1의 결과로 얻은 명도 값에 대해 openCV의 equalizeHist를 이용해 Histogram Equalization 적용
3. openCV의 merge를 이용해 색상, 채도, 명도 값을 묶어서 HSV 형식으로 저장
4. 3의 결과를 RGB 형태로 변환 후에 plot

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
img1 = cv2.imread('chalks_light.tif')
img2 = cv2.imread('stream_dark.tif')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize = (20, 15)) 
rows = 2
cols = 3
  
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(img1)
ax1.set_title('Original')
ax1.axis("off")
 
                               
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(img2)               
ax2.set_title('Original')
ax2.axis("off")
```

```python
hsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
#색상, 채도, 명도
h, s, v = cv2.split(hsv)
equalizedV = cv2.equalizeHist(v)
hsv2 = cv2.merge([h,s,equalizedV])
hsvDst = cv2.cvtColor(hsv2, cv2.COLOR_HSV2RGB)
#hsvDst = cv2.cvtColor(hsvDst, cv2.COLOR_BGR2RGB)

hsv_ = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
#색상, 채도, 명도
h_, s_, v_ = cv2.split(hsv_)
equalizedV_ = cv2.equalizeHist(v_)
hsv_ = cv2.merge([h_,s_,equalizedV_])
hsvDst_ = cv2.cvtColor(hsv_, cv2.COLOR_HSV2BGR)
#hsvDst_ = cv2.cvtColor(hsvDst_, cv2.COLOR_BGR2RGB)
```

```python

fig = plt.figure(figsize = (20, 15)) 
rows = 2
cols = 2
  
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(img1)
ax1.set_title('Original')
ax1.axis("off")
 
                               
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(img2)               
ax2.set_title('Original')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(hsvDst)
ax3.set_title('RGB_Domain_Transform')
ax3.axis("off")
 
                               
ax4 = fig.add_subplot(rows, cols, 4)
ax4.imshow(hsvDst_)               
ax4.set_title('RGB_Domain_Transform')
ax4.axis("off")
```
