```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
I1 = cv2.imread('img1.jpg', cv2.IMREAD_GRAYSCALE)
I2 = cv2.imread('img2.jpg', cv2.IMREAD_GRAYSCALE)

cv2.imshow('image', I1)
cv2.imshow('image2', I2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize = (20, 15)) 
rows = 2
cols = 4
  
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(I1, cmap='gray')
ax1.set_title('Original')
ax1.axis("off")
 
# Step2.첨부된 이미지들의 Fourier Spectrum을 보여주세요.   
f =  np.fft.fft2(I1)                            # 여기 채워넣기, 이미지를 FFT 진행          
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(np.abs(f), cmap='gray')                # 여기 채워넣기
ax2.set_title('After Fourier Transform')
ax2.axis("off")

# Step3. Step2의 결과를 centered shift 시키세요.
f_shift = np.fft.fftshift(f)                            # 여기 채워넣기, FFT된 결과를 shift시켜서 centered FFT 보여주기
ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(np.abs(f_shift), cmap='gray')                # 여기 채워넣기
ax3.set_title('After shifting') 
ax3.axis("off")

# Step4. Step3의 결과를 log transform을 사용해서 더 잘 보이도록 하세요.
magnitude_spectrum = 20 * np.log(np.abs(f_shift))   # 여기 채워넣기, Log transform으로 Spectrum 더 잘 보이게 하기
ax4 = fig.add_subplot(rows, cols, 4)
ax4.imshow(magnitude_spectrum, cmap='gray')
ax4.set_title('After Log Transform')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 5)
ax5.imshow(I2, cmap='gray')
ax5.set_title('Original')
ax5.axis("off")
 
# Step2.첨부된 이미지들의 Fourier Spectrum을 보여주세요.     
f2 =  np.fft.fft2(I2)                                             # 여기 채워넣기
ax6 = fig.add_subplot(rows, cols, 6)
ax6.imshow(np.abs(f2), cmap='gray')                        # 여기 채워넣기
ax6.set_title('After Fourier Transform')
ax6.axis("off")

# Step3. Step2의 결과를 centered shift 시키세요.
f2_shift = np.fft.fftshift(f2)                                        # 여기 채워넣기
ax7 = fig.add_subplot(rows, cols, 7)
ax7.imshow(np.abs(f2_shift), cmap='gray')                      # 여기 채워넣기
ax7.set_title('After shifting')
ax7.axis("off")

# Step4. Step3의 결과를 log transform을 사용해서 더 잘 보이도록 하세요.
magnitude_spectrum2 =  20 * np.log(np.abs(f2_shift))                          # 여기 채워넣기
ax8 = fig.add_subplot(rows, cols, 8)
ax8.imshow(np.abs(magnitude_spectrum2), cmap='gray')                     # 여기 채워넣기
ax8.set_title('After Log Transform')
ax8.axis("off")

plt.show()
```
